```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# importing the dataset into a pandas data frame
card_dataset = pd.read_csv('/content/creditcard.csv')
```

```python
# getting idea about some statistical measures of the dataset
card_dataset.describe()
```

```python
# checking for any missing values in the dataset
card_dataset.isnull().sum()
```

```python
# checking the distribution of fraudulent and legit transactions
card_dataset['Class'].value_counts()

# 0 represents legit transactions
# 1 represents fraudulent transactions
```

```python
# this dataset is highly unbalanced as we can see from the count
# so we'll be separating the data for our analysis

legit = card_dataset[card_dataset.Class == 0.0]
fraud = card_dataset[card_dataset.Class == 1.0]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of our data
legit.Amount.describe()
```

```python
# statistical measures of our data
fraud.Amount.describe()
```

```python
# compare the values for both transactions /
card_dataset.groupby('Class').mean()
```

```python
# instead of dealing with this unbalanced data
# we'll be using Undersampling here

# Build a simple dataset containing similar distribution of legit and fraudulent transactions
# We know the number of fraudulent sampling is 356

legit_sample = legit.sample(n=356)
```

```python
# we did axis = 0 here because we wanted to concat the two dataframes row-wise
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
# so we have a balanced dataset now , we can start training our model
```

```python
new_dataset.groupby('Class').mean()
```

```python
# here we see the dataset parameters for both the datasets are almost the same for both the categories
# this step is important to check whether we got a good or bad sample
```

```python
# now we'll be splitting the features and the target
X = new_dataset.drop(columns=['Class'],axis=1)
Y = new_dataset['Class']
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=2)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
# model making and training
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
# training data accuracy
train_data_prediction = model.predict(X_train)
train_accuracy_score = accuracy_score(Y_train,train_data_prediction)
print(train_accuracy_score)
```

```python
# thats a pretty good accuracy
test_data_prediction = model.predict(X_test)
test_accuracy_score = accuracy_score(Y_test,test_data_prediction)
print(test_accuracy_score)
```
